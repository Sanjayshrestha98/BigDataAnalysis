```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
import pandas as pd
import matplotlib.pyplot as plt

spark = SparkSession.builder \
    .appName("EnergyAnalysis") \
    .config("spark.sql.shuffle.partitions", "4") \
    .getOrCreate()
spark.sparkContext.setLogLevel("WARN")
```

```python
# Handling missing values
datacheck = df.na.drop(subset=["coal_prod_change_pct","coal_cons_change_pct"])

# datacheck.select("coal_prod_change_pct").show()
average_coalprod_per_year = datacheck.groupBy("year").avg("coal_prod_change_pct")
average_coalCons_per_year = datacheck.groupBy("year").avg("coal_cons_change_pct")

result_df = average_coalprod_per_year.withColumnRenamed('avg(coal_prod_change_pct)', 'avg_coal_prod_change_pct').join(
    average_coalCons_per_year.withColumnRenamed('avg(coal_cons_change_pct)', 'avg_coal_cons_change_pct'),
    on="year",
    how="inner" 
)



```

```python
# Removing outliers
outlier_columns = ["avg_coal_prod_change_pct", "avg_coal_cons_change_pct"]

lower_bounds = {}
upper_bounds = {}
for column in outlier_columns:
    q1, q3 = result_df.approxQuantile(column, [0.5, 0.95], 0.01)
    iqr = q3 - q1
    lower_bounds[column] = q1 - 1.5 * iqr
    upper_bounds[column] = q3 + 1.5 * iqr

filtered_df = result_df
for column in outlier_columns:
    filtered_df = filtered_df.filter((col(column) >= lower_bounds[column]) & (col(column) <= upper_bounds[column]))

filtered_df.show()
```

```python
# Convert PySpark DataFrame to pandas DataFrame for plotting
pandas_result_df = filtered_df.orderBy('year').toPandas()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(pandas_result_df['year'], pandas_result_df['avg_coal_prod_change_pct'], label='Average Coal Production')
plt.plot(pandas_result_df['year'], pandas_result_df['avg_coal_cons_change_pct'], label='Average Coal Consumption')
plt.xlabel('Year')
plt.ylabel('Average Change Percentage')
plt.title('Average Coal Production and Consumption Change Over Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Handling missing values
datacheck = df.na.drop(subset=["coal_prod_change_pct","country","fossil_cons_change_pct", "gdp",'fossil_cons_per_capita', 'population'])
# df.na.fill(value=0,subset=["population"]).show()
selected_columns = ["iso_code", "year", "coal_prod_change_pct", "country","fossil_cons_change_pct", "gdp", "fossil_cons_per_capita", 'population']

# Modifying columns
avg_fossil_cons_per_capita = datacheck.select(selected_columns).groupBy("country").avg('fossil_cons_per_capita')

avg_fossil_cons_per_capita_year = datacheck.select(selected_columns).groupBy("year").avg('fossil_cons_per_capita')

avg_population = datacheck.select(selected_columns).groupBy("country").avg('population')

avg_coal_change = datacheck.select('year')

avg_values = avg_fossil_cons_per_capita.withColumnRenamed('avg(fossil_cons_per_capita)', 'avg_fossil_cons_per_capita').join(
    avg_population.withColumnRenamed('avg(population)', 'avg_population'),
    on="country",
    how="inner"  # You can choose the appropriate join type
)

avg_values.show()
```

```python

outlier_columns = ["avg_population", "avg_fossil_cons_per_capita"]

lower_bounds = {}
upper_bounds = {}
for column in outlier_columns:
    q1, q3 = avg_values.approxQuantile(column, [0.25, 0.75], 0.01)
    iqr = q3 - q1
    lower_bounds[column] = q1 - 1.5 * iqr
    upper_bounds[column] = q3 + 1.5 * iqr

filtered_df = avg_values
for column in outlier_columns:
    filtered_df = filtered_df.filter((col(column) >= lower_bounds[column]) & (col(column) <= upper_bounds[column]))

filtered_df.show()

```

```python
# Select relevant columns for analysis
selected_columns_for_clustering = ["country", "fossil_cons_change_pct", 'coal_prod_change_pct', "gdp"]
avg_fossil_cons_rate = datacheck.select(*selected_columns_for_clustering).groupBy('country').avg('fossil_cons_change_pct')

avg_coal_prod_rate = datacheck.select(*selected_columns_for_clustering).groupBy('country').avg('coal_prod_change_pct')
avg_coal_prod_rate.show()
```

```python
from pyspark.ml.clustering import KMeans
from pyspark.ml.feature import StandardScaler
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression

feature_columns = ["avg_fossil_cons_per_capita", 'avg_population']

assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")
data_assembled = assembler.transform(filtered_df)

# Standardize features
scaler = StandardScaler(inputCol="features", outputCol="scaled_features")
data_scaled = scaler.fit(data_assembled).transform(data_assembled)

kmeans = KMeans(k=10, seed=1, featuresCol="scaled_features", predictionCol="cluster")
model = kmeans.fit(data_scaled)

# Assign clusters to data points
clustered_data = model.transform(data_scaled)

# Show the resulting clusters
# clustered_data.select("country", "fossil_fuel_consumption", "cluster").show()

clustered_data.show()
```

```python
import matplotlib.pyplot as plt

# Convert PySpark DataFrame to Pandas DataFrame
pandas_clustered_data = clustered_data.toPandas()

# Scatter plot to visualize clusters
plt.scatter(
    pandas_clustered_data["avg_fossil_cons_per_capita"],
    pandas_clustered_data["avg_population"],
    c=pandas_clustered_data["cluster"],
    cmap="viridis", 
    s=50  # Marker size
)
plt.xlabel("Average Fossil Consumption per Capita")
plt.ylabel("Average Population")
plt.title("Clustering Visualization")
plt.colorbar(label="Cluster")
plt.show()
```

```python
# REGRESSION

# Select relevant columns
selected_columns = ["fossil_fuel_consumption", "gdp"]
feature_columns =["gdp"]
df_selected = df.select(selected_columns).na.drop()

df_selected.show()
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import RegressionEvaluator

# Prepare the features (independent variables) and the target variable (dependent variable)
assembler = VectorAssembler(inputCols=selected_columns, outputCol="features")
assembled_data = assembler.transform(df_selected)
```

```python

# Split the data into training and testing sets
train_data, test_data = assembled_data.randomSplit([0.7, 0.3], seed=42069)

# Create a LinearRegression model
lr = LinearRegression(featuresCol="features", labelCol="fossil_fuel_consumption")
```

```python

# Build a pipeline
pipeline = Pipeline(stages=[lr])

# Fit the model on the training data
model = pipeline.fit(train_data)

```

```python
# Make predictions on the test data
predictions = model.transform(test_data)

# Display the predicted values and actual values
predictions.select("prediction", "fossil_fuel_consumption", "gdp").show()

```

```python

# Evaluate the model using RMSE
evaluator = RegressionEvaluator(labelCol="fossil_fuel_consumption", predictionCol="prediction", metricName="rmse")
rmse = evaluator.evaluate(predictions)
print("Root Mean Squared Error (RMSE):", rmse)
```
